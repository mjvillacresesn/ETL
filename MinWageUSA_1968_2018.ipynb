# Data Sets (source: dol.gov)

```python
# In this case we can use pandas_to_html to parse tables from website into a pandas dataframe

# 4. Use pandas 'read_html' to abtract tables from web by identifying each table with its specific summary attribute
#1968-1981 
data1 = pd.read_html('https://www.dol.gov/whd/state/stateMinWageHis.htm', attrs={'summary': 'This is a table of minimum changes by state for 1968 to 1981.'})[0]
df1 = data1.reindex(data1.index.drop(0))
df1.head(3)
```

```python
# 7. Repeat process for all 5 tables
#1988-1998
data2 = pd.read_html('https://www.dol.gov/whd/state/stateMinWageHis.htm',attrs={'summary': 'This is a table of minimum changes by state for 1988 to 1998.'})[0]
df2 = data2.reindex(data2.index.drop(0))
df2.head(3)
```

```python
#2000-2006
data3 = pd.read_html('https://www.dol.gov/whd/state/stateMinWageHis.htm',attrs={'summary': 'This is a table of minimum changes by state for 2000 to 2006.'})[0]
df3 = data3.reindex(data3.index.drop(0))
df3.head(3)
```

```python
#2007-2013
data4 = pd.read_html('https://www.dol.gov/whd/state/stateMinWageHis.htm',attrs={'summary': 'This is a table of minimum changes by state for 2007 to 2013.'})[0]
df4 = data4.reindex(data4.index.drop(0))
df4.head(3)
```

```python
#2014-2018
data5 = pd.read_html('https://www.dol.gov/whd/state/stateMinWageHis.htm',attrs={'summary': 'This is a table of minimum changes by state for 2014 to 2017.'})[0]
df5 = data5.reindex(data5.index.drop(0))
df5.head(3)
```

# Merge Tables

```python
# 8. Merge all 5 dfs on "State or other  jurisdiction"; dates (1968-2018). Note: Be mindful of spacing
df1.merge(df2,on='State or other  jurisdiction').merge(df3,on='State or other  jurisdiction').merge(df4,on='State or other  jurisdiction').merge(df5,on='State or other  jurisdiction')
```
